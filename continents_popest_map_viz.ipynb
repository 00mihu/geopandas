```python
import matplotlib.pyplot as plt
```

```python
f, ax = plt.subplots(1,1,figsize=(10,6))
gdf.plot(ax=ax, column = 'continent', cmap = 'tab10', edgecolor = 'w')
ax.axis('off')
```

```python
from matplotlib import pyplot as plt
import seaborn as sns
gdf.groupby('continent').size().plot(kind='barh', color=sns.palettes.mpl_palette('Dark2'))
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
from matplotlib import pyplot as plt
import seaborn as sns
figsize = (12, 1.2 * len(gdf['continent'].unique()))
plt.figure(figsize=figsize)
sns.violinplot(gdf, x='pop_est', y='continent', inner='stick', palette='Dark2')
sns.despine(top=True, right=True, bottom=True, left=True)
```
