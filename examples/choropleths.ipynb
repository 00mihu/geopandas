```python
import geopandas as gp
import pysal as ps 
```

```python
pth = ps.examples.get_path("columbus.shp")
tracts = gp.GeoDataFrame.from_file(pth)
```

```python
tracts.plot(column='CRIME', scheme='QUANTILES', k=4, colormap='OrRd')
```

```python
tracts.plot(column='CRIME', scheme='Equal_Interval', colormap='OrRd')
```

```python
tracts.plot(column='CRIME', scheme='Fisher_Jenks', colormap='OrRd')
```

```python
tracts.plot(column='CRIME', scheme='Fisher_Jenks', k=7, colormap='OrRd')
```
